```python
import keras
import pandas as pd
from keras.models import Sequential
from keras.models import Model
from keras.layers import *
import tensorflow as tf
```

```python
csv_train = 'trainPredMaint.csv'
csv_test = 'testPredMaint.csv'
train_data = pd.read_csv(csv_train)
test_data = pd.read_csv(csv_test)
```

```python
print(train_data.columns)
```

```python
print(test_data.columns)
```

```python
drop_columns = ['UnitNumber','Time', 'Setting1', 'Setting2', 'Setting3']
X = train_data.drop(drop_columns + ['RUL'], axis=1)
print(X.columns)
Y = train_data['RUL']
```

```python
input_size = X.columns.size

model = Sequential()
model.add(Dense(25, input_dim=input_size, activation='relu',name='layer_1'))
model.add(Dense(12, activation='relu', name='layer_2'))
model.add(Dense(25, activation='relu',name='layer_3'))
model.add(Dense(1, activation='linear', name='output_layer'))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X,Y, epochs=100, shuffle=True, verbose=1)
```

```python
model.save('kerasiq.mdl')
```

```python
error = model.evaluate(X,Y)
print(np.sqrt(error))
```

```python
X_test = test_data.drop(drop_columns, axis=1)
prediction = model.predict(X_test)
len(prediction)
```

### autoencoding

```python
input_size = 21
hidden_size = 15
X_auto = Input(shape=(input_size,))
# Encoder
h = Dense(hidden_size, activation='relu')(X_auto)
# Decoder
r = Dense(input_size, activation='sigmoid')(h)

autoencoder = Model(input=X_auto, output=r)
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
history = autoencoder.fit(X, X, epochs=5, verbose=1)
```
